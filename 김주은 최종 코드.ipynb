```python
# 확장형 무인 주문 시스템

menu = {
    1: ["아메리카노", 3000],
    2: ["카페라떼", 4000],
    3: ["카푸치노", 4500],
    4: ["바닐라라떼", 4800]
}

cart = []  # 장바구니


def show_menu():
    """메뉴 출력 함수"""
    print("\n====== 메뉴 ======")
    for num, info in menu.items():
        print(f"{num}. {info[0]} - {info[1]}원")
    print("0. 주문 완료")


def order_menu():
    """메뉴 선택 + 수량 입력을 여러 번 처리하는 함수"""
    while True:
        try:
            choice = int(input("메뉴 번호 선택: "))

            if choice == 0:
                break

            if choice not in menu:
                print("잘못된 메뉴 번호입니다.")
                continue

            quantity = int(input("수량 입력: "))

            if quantity <= 0:
                print("수량은 1개 이상이어야 합니다.")
                continue

            cart.append([menu[choice][0], menu[choice][1], quantity])
            print(f"{menu[choice][0]} {quantity}개 추가 완료")

        except ValueError:
            print("숫자만 입력하세요.")


def show_order():
    """주문 내역 및 총 금액 출력 함수"""
    print("\n====== 주문 내역 ======")
    total = 0

    for name, price, quantity in cart:
        cost = price * quantity
        total += cost
        print(f"{name} x {quantity} = {cost}원")

    print("----------------------")
    print(f"총 결제 금액: {total}원")


def main():
    """메인 함수"""
    print("무인 주문 시스템 시작")
    show_menu()
    order_menu()

    if cart:
        show_order()
    else:
        print("주문 내역이 없습니다. 다시 시도 하십시오.")

    print("주문 종료")


# 프로그램 실행
main()
```
